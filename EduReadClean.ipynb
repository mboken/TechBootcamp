After closely reading the problem statement on driven data, I came to understand that the columns "Function" through "Operating Status" each contain a label.  The remaining columns are all freform text data (aside from FTE and Total, which are freeform float values).

Although the columns containing a label sometimes contain "NO LABEL," we decided that we didn't want to drop these rows or replace "NO LABEL" with NA. The "NO LABEL" gives us insights that the label didn't fall into one of the other categories listed.

```python
# After inspecting Text_ 1-4, it was clear that the information in each of these columns did not fall under
# specific categories and could all be contained in one column

# First, fill na values with ' ' to make addition of columns possible
education[["Text_1","Text_2","Text_3","Text_4"]]= education[["Text_1","Text_2","Text_3","Text_4"]].fillna(' ')
education["Text_1"] = education["Text_1"] + education["Text_2"] + education["Text_3"] + education["Text_1"]
```

```python
# Check to make sure it worked:
pd.isnull(education["Text_1"]).sum()
```

```python
# Now drop Text_2-4
education.drop(columns=["Text_2","Text_3","Text_4"],inplace=True)
```

```python
# To deal with NA values, we can impute by utilizing the most common value for each column

# build a function to generate most common value for a given column:
def most_frequent(column_name):
    # pull all values for that label in function and the column for which we want to find the most frequent value
    all_values = education[column_name]
    # initialize dictionary
    counter = {}
    # fill dictionary with key value pairs to find the most common entry
    for word in all_values:
        if not pd.isnull(word):
            if word in counter:
                counter[word]+=1
            else:
                counter[word]=1
    Keymax = max(counter, key=counter.get)
    # return entry that appeared most frequently
    return Keymax
```

```python
# initialize dictionary to state most frequent value per column
frequent_dict = {}

# Create list of freeform text columns
freeform_columns = education.columns[9:]

# Populate the dictionary
for col in freeform_columns:
    frequent_dict[col] = most_frequent(col)
```

```python
#Check the frequent_dict to make sure it worked:
frequent_dict
```

```python
# Now impute using fillna:
education.fillna(value=frequent_dict,inplace=True)
```

```python
# double check that all NA values have been replaced
pd.isnull(education).sum()
```
